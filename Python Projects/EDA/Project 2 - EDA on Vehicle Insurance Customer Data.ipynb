### Table1 ( customer_details )

```python
#importing customer_details.csv
df1=pd.read_csv('customer_details.csv')
df1
```

```python
# Column Name for customer details table:
df1.columns=['customer_id', 'Gender','age',  'driving_licence_present','region_code','previously_insured','vehicle_age','vehicle_damage']
df1
```

```python
# checking and cleaning data from customer_details.csv
df1.info()
```

```python
# Null values of columns
# Generate a summary of count of all the null values column wise

print('Null Values in Cusmoter_id:',df1['customer_id'].isna().sum())
print('Null Values in Gender:',df1['Gender'].isna().sum())
print('Null Values in age:',df1['age'].isna().sum())
print('Null Values in driving_licence_present:',df1['driving_licence_present'].isna().sum())
print('Null Values in region_code:',df1['region_code'].isna().sum())
print('Null Values in previously_insured:',df1['previously_insured'].isna().sum())
print('Null Values in vehicle_age:',df1['vehicle_age'].isna().sum())
print('Null Values in vehicle_damage:',df1['vehicle_damage'].isna().sum())
```

```python
#droping and filling null values
#Drop Null values for customer_id because central tendencies for idâ€™s is not feasible.

df1.dropna(subset=['customer_id'])
df1.dropna(inplace=True)
print('Null Values in Cusmoter_id after droping Null values:',df1['customer_id'].isna().sum())
```

```python
#Replace all null values for numeric columns by mean. 

df1['age'].fillna(df1['age'].mean(),inplace=True)
print('Null Values in age after droping Null values:',df1['age'].isna().sum())
```

```python
# Replace all null values for Categorical value by mode.

df1['driving_licence_present'].fillna(df1['driving_licence_present'].mode()[0],inplace=True)
print('Null Values in driving_licence_present after droping Null values:',df1['driving_licence_present'].isna().sum())
 
df1['region_code'].fillna(df1['region_code'].mode()[0],inplace=True)
print('Null Values in region_code after droping Null values:',df1['region_code'].isna().sum())
 
df1['previously_insured'].fillna(df1['previously_insured'].mode()[0],inplace=True)
print('Null Values in previously_insured after droping Null values:',df1['previously_insured'].isna().sum())

df1['customer_id'].fillna(df1['customer_id'].mode()[0],inplace=True)
print('Null Values in Cusmoter_id after droping Null values:',df1['customer_id'].isna().sum())

df1['Gender'].fillna(df1['Gender'].mode()[0],inplace=True)
print('Null Values in Gender after droping Null values:',df1['Gender'].isna().sum())
 
df1['vehicle_age'].fillna(df1['vehicle_age'].mode()[0],inplace=True)
print('Null Values in vehicle_age after droping Null values:',df1['vehicle_age'].isna().sum())
 
df1['vehicle_damage'].fillna(df1['vehicle_damage'].mode()[0],inplace=True)
print('Null Values in vehicle_damage after droping Null values:',df1['vehicle_damage'].isna().sum())
```

### Table 2  ( customer_policy_details )

```python
df2 = pd.read_csv("customer_policy_details.csv")
df2
```

```python
# Column Name for customer_policy table:
df2.columns=['customer_id','annual_premium(Rs)','sales_channel_code','vintage','response']

df2
```

```python
#checking and cleaning data from customer_details.csv
df2.info()
```

```python
# Null values of columns
# Generate a summary of count of all the null values column wise

print('Null Values in Cusmoter_id:',df2['customer_id'].isna().sum())
print('Null Values in annual_premium(Rs):',df2['annual_premium(Rs)'].isna().sum())
print('Null Values in sales_channel_code:',df2['sales_channel_code'].isna().sum())
print('Null Values in vintage:',df2['vintage'].isna().sum())
print('Null Values in response:',df2['response'].isna().sum()) 
```

```python
#droping and filling null values
#Drop Null values for customer_id because central tendencies for idâ€™s is not feasible.
df2.dropna(subset=['customer_id'])
df2.dropna(inplace=True)
print('Null Values in Cusmoter_id after droping Null values:',df2['customer_id'].isna().sum())
```

```python
#Replace all null values for numeric columns by mean.

df2['annual_premium(Rs)'].fillna(df2['annual_premium(Rs)'].mean(),inplace=True)
print('Null Values in annual_premium(Rs) after droping Null values:',df2['annual_premium(Rs)'].isna().sum())
 
df2['sales_channel_code'].fillna(df2['sales_channel_code'].mean,inplace=True)
print('Null Values in sales_channel_code after droping Null values:',df2['sales_channel_code'].isna().sum())
 
df2['vintage'].fillna(df2['vintage'].mean,inplace=True)
print('Null Values in vintage after droping Null values:',df2['vintage'].isna().sum())
 
df2['response'].fillna(df2['response'].mean,inplace=True)
print('Null Values in response after droping Null values:',df2['response'].isna().sum())
```

```python
df2.describe()
```

```python
# Generate a summary of count of all the outliers column wise
# Replace all outlier values for numeric columns by mean. 

Q1=df2.describe().loc["25%","annual_premium(Rs)"]
Q3=df2.describe().loc["75%","annual_premium(Rs)"]

IQR=Q3-Q1
print("IQR value of annual_premium(Rs)",IQR)

print('Outliners in annual_premium(Rs):',(Q1-IQR*1.5))
print('Outliners in annual_premium(Rs):',(Q3+IQR*1.5))

print('Outliners in annual_premium(Rs):',df2.loc[df2["annual_premium(Rs)"]<(Q1-IQR*1.5),"annual_premium(Rs)"].count())
print('Outliners in annual_premium(Rs):',df2.loc[df2["annual_premium(Rs)"]>(Q3+IQR*1.5),"annual_premium(Rs)"].count())
```

```python
a = (df2[["annual_premium(Rs)"]]<(Q1-IQR*1.5)) | (df2[["annual_premium(Rs)"]]>(Q3+IQR*1.5))
b = a[a["annual_premium(Rs)"]].index
b
```

```python
df2.loc[b,"annual_premium(Rs)"]=df2["annual_premium(Rs)"].mean()
```

```python
Q1=df2.describe().loc["25%","vintage"]
Q3=df2.describe().loc["75%","vintage"]

IQR=Q3-Q1
print("IQR value of vintage",IQR)

print('Outliners in vintage:',(Q1-IQR*1.5))
print('Outliners in vintage:',(Q3+IQR*1.5))

print('Outliners in vintage:',df2.loc[df2["vintage"]<(Q1-IQR*1.5),"vintage"].count())
print('Outliners in vintage:',df2.loc[df2["vintage"]>(Q3+IQR*1.5),"vintage"].count())
```

### Table 1 ( customer_details ) and Table 2 ( customer_policy_details ) merge data

```python
#Merging
master_data=pd.merge(df1,df2,on='customer_id')
master_data
```

```python
#Gender wise average annual premium

Gender_wise_average_annual_premium = master_data.groupby(["Gender"])["annual_premium(Rs)"].mean()
Gender_wise_average_annual_premium
```

```python
import matplotlib.pyplot as pyplot
Gender_wise_average_annual_premium.plot()
pyplot.show()
```

```python
#Age wise average annual premium

Age_wise_average_annual_premium = master_data.groupby(["age"])["annual_premium(Rs)"].mean()
Age_wise_average_annual_premium
```

```python
import matplotlib.pyplot as pyplot
Age_wise_average_annual_premium.plot(color='r')
pyplot.show()
```
