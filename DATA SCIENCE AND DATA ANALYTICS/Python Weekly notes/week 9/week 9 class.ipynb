```python
# merging DataFrames
emp = pd.DataFrame({
    'empno':[101,102,103,104,105],
    'ename':['Vinay','Pravalika','Hema','Hari','Ashok'],
    'deptno':[1,2,2,1,3]
})
dept= pd.DataFrame({
    'deptno':[1,2,3],
    'dname':['Marketing','Operations','EDP']
})
merge_df = pd.merge(emp,dept,on='deptno') #key is deptno
merge_df
```

```python
emp = pd.DataFrame({
    'empno':[324,512,361,423,206,516,105],
    'ename':['Vinay','pravalika','Purushottam','Hari','Hema','jayath','Ashok'],
    'deptno':[1,5,1,3,4,2,5]
})
dept= pd.DataFrame({
    'deptno':[1,2,3,4,5,6],
    'dname':['Mechanical','Civil','Electrical','Electronic','Computer Science','Data science']
})
merge_df = pd.merge(emp,dept,on='deptno') #key is deptno
merge_df
```

```python
merge_left = pd.merge(emp,dept,on = 'deptno',how = 'left')
merge_left
```

```python
merge_right = pd.merge(emp,dept,on = 'deptno',how = 'right')
merge_right
```

```python
merge_outer = pd.merge(emp,dept,on = 'deptno',how = 'outer')
merge_outer
```

```python
merge_inner = pd.merge(emp,dept,on = 'deptno',how = 'inner')
merge_inner
```
