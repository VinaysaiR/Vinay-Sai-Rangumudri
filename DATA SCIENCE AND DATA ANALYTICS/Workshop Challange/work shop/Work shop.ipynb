#### Prepare the Data
#### We will now use pd to datetime() function from the Pandas module. The input to the pd to datetime() function is the Pandas series whose values are to be converted into datetime values
#### Syntax: pd to datetime(series)
#### The default format of a datetime value is YYYY-MM-DD.

```python
df1["Date"]= pd.to_datetime(df1['Date'],format= '%d-%m-%Y')
df1.info()
```

```python
#extracting date series
date_series=df1["Date"]
date_series
```

```python
#extracting year from date
year_series = date_series.dt.year 
year_series
```

```python
#extracting month and day
month_series = date_series.dt.month 
day_series = date_series.dt.day
```

```python
df1['year'] = year_series
df1['month'] = month_series
df1['day'] = day_series
df1
```

```python
#Check for the missing values in the 'df' DataFrane. 
df1.isnull().sum()
```

```python
# Check for the missing values in the 'df' DataFrame.
df1.isnull()
```

#### More than 10000 values contained in the PM10, NH3,Xylene columns are the missing (or null) values. So, let's drop these columns using the drop() function. It removes rows or columns by specifying lebel names and corresponding axis. We will specify the column names

```python
#Drop the PM10, NH3, Xylene columns from the 'df' DataFrame.
df2 = df1.drop(columns=['PM10', 'NH3', 'Xylene'], axis = 1 )
df2
```

```python
#Get the List of columns present in the 'df' DataFrame after removing the PM10, NH3, Xylene columns.
df2.columns
```

```python
# Lets Check again for the missing values in the 'of' DataFrame.
df2.isnull().sum()
```

```python
df2.info()
```

#### We know that the other columns have missing values. So let's see which rows in the PM2.5 column are having missing values

```python
df2[df2["PM2.5"].isnull() == True ]
```

```python
#replace null values in each column
for col in df2.columns[2:-3]:
    row_ind = df2[df2[col].isnull() == True].index
    df2.loc[row_ind,col] = df2[col].median()
```

```python
#Check for missing values again 
df2.isnull().sum()
```

#### Analyze the Data - Data Transformation
#### Group/Aggregation functions are the most commonly used functions for analysing

```python
#Data frame with data about the year 2015. 
df_2015 = df2.loc[df2['year'] == 2015, df2.columns]
df_2015
```

```python
#Data frame with data about the year 2016. 
df_2016 = df2.loc[df2['year'] == 2016, df2.columns]
df_2016
```

```python
# Data frames for the year 2017, 2018, 2019, 2020.
df_2017 = df2.loc[df2['year'] == 2017, df2.columns]
df_2018 = df2.loc[df2['year'] == 2018, df2.columns]
df_2019 = df2.loc[df2['year'] == 2019, df2.columns]
df_2020 = df2.loc[df2['year'] == 2020, df2.columns]
```

```python
#Lets group the data frame year wise using groupby() and get the data for 2020 year 
gp_year = df2.groupby(by='year') 
yr_2019 = gp_year.get_group(2019) 
yr_2020 = gp_year.get_group(2020)
yr_2020
```

```python
yr_2019
```

```python
# Apply mean and median for all the years using agg().
gp_year.agg(func=['mean', 'median'])
```

```python
# Find median of CO in each year 
gp_year[['CO']].median()
```

```python
#Group the data by month using groupby() 
gp_month = df2.groupby(by='month')
gp_month.get_group(12)
```

```python
gp_month ['PM2.5'].median()
```

```python
# Apply different functions on different columns using dictionary in the agg()
gp_month.agg (func={'PM2.5': ['min', 'max'], 'month': 'count'})
#month_min_max =
#month_min_max
```
